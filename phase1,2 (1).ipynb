```python
import pandera as pa
from pandera import Column, DataFrameSchema, Check, Index

# Define the schema with more nullable options
schema = pa.DataFrameSchema({
    "Formatted Date": Column(str),
    "Summary": Column(str),
    "Precip Type": Column(str, nullable=True),  # This might have nulls
    "Temperature (C)": Column(float, nullable=True),  # Allow null values
    "Apparent Temperature (C)": Column(float, nullable=True),  # Allow null values
    "Humidity": Column(float),
    "Wind Speed (km/h)": Column(float, nullable=True),  # This might have nulls
    "Wind Bearing (degrees)": Column(int),
    "Visibility (km)": Column(float , nullable=True) ,
    "Loud Cover": Column(int),
    "Pressure (millibars)": Column(float),
    "Daily Summary": Column(str)
})

# Load the data
weather_df = pd.read_csv("/content/weatherHistory.csv", encoding='utf-8')

# Validate the dataframe
validated_df = schema.validate(weather_df)
print(validated_df)
```

```python
import string
import random

# define function that operates on entire dataframe
def randomize_values(df):
    for column in df.columns:
        if df[column].dtype == 'O': # check if column has object dtype
            df[column] = [''.join(random.choices(string.ascii_letters + string.digits, k=10))
            for _ in range(len(df))] # generate a list of random strings
    return df
```

```python
# apply function to dataframe
df_rand = randomize_values(df)

df_rand
```

```python
import numpy as np
import pandas as pd

# Function to anonymize float columns
def anonymize_floats(column, scale=0.05):
    noise = np.random.normal(0, scale, size=len(column))
    return column + noise

# Anonymize specific float columns
df['Temperature (C)'] = anonymize_floats(df['Temperature (C)'])
df['Humidity'] = anonymize_floats(df['Humidity'])
df['Wind Speed (km/h)'] = anonymize_floats(df['Wind Speed (km/h)'])
df['Visibility (km)'] = anonymize_floats(df['Visibility (km)'])
df['Pressure (millibars)'] = anonymize_floats(df['Pressure (millibars)'])

# Display the anonymized DataFrame
print(df)
```

# perturbation

```python
import numpy as np

dfp = pd.read_csv('/content/weatherHistory.csv')


def add_noise(dfp, column, std = None):   #std=standaard deviation
    if std == None:
        std = dfp[column].std()

    withNoise = dfp[column].add(np.random.normal(0, std, dfp.shape[0]))
    copy = dfp.copy()
    copy[column] = withNoise
    return copy

perturbation = add_noise(dfp, 'Temperature (C)', std=100)
perturbation
```

```python
import nltk  #Natural Language Toolkit (NLTK)
from nltk.tokenize import word_tokenize
df_token= pd.read_csv('/content/weatherHistory.csv')


# Download punkt tokenizer
nltk.download('punkt') #pre-trained unsupervised machine learning model provided by NLTK for tokenizing text into sentences.


# Define a function to tokenize data in a specific column
def tokenize_data(data):
   return word_tokenize(data)


df_token['Tokenized'] = df_token['Daily Summary'].apply(tokenize_data)

df_token
```

# Data Psudonomization

```python
# Define a function to scramble data in a specific column
def scramble_data(data):
    data_list = list(data)
    random.shuffle(data_list)
    return ''.join(data_list)


df['Scrambled_description'] = df['Daily Summary'].apply(scramble_data)

# Display the result
print(df[['Daily Summary', 'Scrambled_description']].head())
```

```python
import hashlib

# Hash the 'Treatment' column using a lambda function
df['Hashed_Summary'] = df['Summary'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())

# Display the result
print(df[['Summary', 'Hashed_Summary']])
```
